### 2. ทดสอบ independence of errors
2.1 ตรวจสอบโดยการ plot time vs residuals (ถ้าสามารถทำได้)

2.2 ตรวจสอบด้วย Durbin-Wutson test  

H0: There is no correlation among the residuals.  
H1: The residuals are correlated.

The Durbin-Watson statistic ranges in value from 0
to 4. A value near 2 indicates non-correlation; a value toward 0 indicates positive
correlation; a value toward 4 indicates negative correlation.

Reject H0 if d < dL  
Accept H0 if d > dU

Durbin-Watson table: https://www.real-statistics.com/statistics-tables/durbin-watson-table/

```python
auto_lr2 = smf.ols('mpg ~ horsepower+I(horsepower**2)', data=auto).fit()
print(auto_lr2.summary())
```

### ดูตาราง Durbin-Watson Alpha = .05 n = 350 และ k = 2

ได้ค่า dL = 1.813 dU = 1.836

ดังนั้น Reject H0 -- มี correlation ระหว่าง residuals

### 3. ทดสอบ constant variance

### ใช้ Breush-Pagan test


H0: The residuals are distributed with equal variance  
H1: The residuals are not distributed with equal variance


```python
import statsmodels.stats.api as sms

name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = sms.het_breuschpagan(auto_lr2.resid, auto_lr2.model.exog)
pd.DataFrame({'Name':name, 'value':test})
```

### p-value < 0.05 ดังนั้น accept H1

### Bad = Intercept + other values
### Medium = Intercept + other values + ShelveLoc(Med)*1
### Good = Intercept + other values + ShelveLoc(Good)*1

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan
import numpy as np
import matplotlib.pyplot as plt


carseats['ShelveLoc'] = pd.Categorical(carseats['ShelveLoc'], categories=['Bad','Medium','Good'], ordered=True)

df = carseats.dropna().copy()
formula = 'Sales ~ CompPrice + Income + Advertising + Population + Price + C(ShelveLoc) + Age + Education + C(Urban) + C(US)'
model = smf.ols(formula, data=df).fit()
print(model.summary())

# -----------------------------
# 1,2
b_med   = model.params.get('C(ShelveLoc)[T.Medium]', np.nan)
b_good  = model.params.get('C(ShelveLoc)[T.Good]',   np.nan)
diff_gm = b_good - b_med

print('\n[ข้อ 1] Medium เทียบ Bad =', b_med)
print('[ข้อ 2] Good เทียบ Medium =', diff_gm, '  (เพราะ Good-Bad ลบ Medium-Bad)')

# -----------------------------
#ข้อ 3  3.1, 3.3, 3.4, 3.5
fitted = model.fittedvalues
resid  = model.resid

plt.figure()
plt.scatter(fitted, resid)
plt.axhline(0, linestyle='--')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()

# QQ plot
sm.qqplot(resid, line='45', fit=True)
plt.title('QQ plot of residuals')
plt.show()

# Scale-Location plot
from statsmodels.stats.outliers_influence import OLSInfluence
infl = OLSInfluence(model)
std_resid = infl.resid_studentized_internal

plt.figure()
plt.scatter(fitted, np.sqrt(np.abs(std_resid)))
plt.xlabel('Fitted values')
plt.ylabel('Sqrt(|Standardized residuals|)')
plt.title('Scale-Location plot')
plt.show()

# Influence plot
sm.graphics.influence_plot(model, criterion="cooks")
plt.title('Influence plot (leverage vs studentized residuals)')
plt.show()

# -----------------------------
# ตัวเลขประกอบข้อ 3

# 3.2 Independence of errors → ใช้ค่า Durbin–Watson จาก summary
from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(resid)
print('\nDurbin–Watson =', round(dw,3), '(~2 = อิสระดี)')

# 3.3 Constant variance (Homoskedasticity) → Breusch–Pagan test
bp_test = het_breuschpagan(resid, model.model.exog)
bp_LM, bp_pvalue, fvalue, fpvalue = bp_test
print('Breusch–Pagan p-value =', fpvalue, '(p>0.05 → ความแปรปรวนคงที่)')

# 3.4 Outliers → Bonferroni outlier test (studentized residuals)
out_test = model.outlier_test()
sus = out_test[out_test['bonf(p)'] < 0.05]
print('\nจุดต้องสงสัยเป็น outlier (Bonferroni p<0.05):')
print(sus)

# 3.5 Leverage points → ดู leverage
n  = df.shape[0]
p  = len(model.params)-1
lev = infl.hat_matrix_diag
lev_cut = 2*(p+1)/n
sus_lev_idx = np.where(lev > lev_cut)[0]
print(f'\nเกณฑ์ leverage cutoff ≈ {lev_cut:.3f}; พบจุด leverage สูง {len(sus_lev_idx)} จุด (index ในตาราง):')
print(sus_lev_idx[:20])

# 3.6 Multicollinearity → VIF
X = model.model.exog
vif = pd.DataFrame()
vif['variable'] = model.model.exog_names
vif['VIF'] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print('\nVIF:')
print(vif)
```
